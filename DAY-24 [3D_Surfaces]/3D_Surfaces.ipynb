# 3D_Surfaces

In data visualization and mathematical computations, a meshgrid is a method used to create a grid of coordinate points in a multi-dimensional space, typically for the purpose of evaluating functions or creating visual representations of data. Meshgrids are particularly useful in two-dimensional (2D) and three-dimensional (3D) plotting scenarios.

Here's what a meshgrid signifies and how it is used:

Coordinate Grid: A meshgrid represents a grid of coordinate points in a specified range for each dimension. For example, in 2D, you have an x-coordinate grid and a y-coordinate grid. In 3D, you would have x, y, and z-coordinate grids.

Evaluation Points: These grids of coordinate points are used to define a set of points where you can evaluate a function. For each point in the grid, you can calculate the value of a function or retrieve data. This is particularly useful when visualizing functions of multiple variables.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

X,Y = np.meshgrid(x,y)
X.shape

Z = X**2 + Y**2


fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(X,Y,Z,cmap='viridis')
fig.colorbar(p)

plt.show()
```

```python
Z = np.sin(X) + np.sin(Y)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(X,Y,Z,cmap='cool')
fig.colorbar(p)

plt.show()
```

```python
# contour represent 3D graph into 2D

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

p = ax.contour(X,Y,Z,cmap='viridis')
fig.colorbar(p)

plt.show()
```

```python
# contour represent 3D graph into 2D

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

p = ax.contourf(X,Y,Z,cmap='viridis')
fig.colorbar(p)

plt.show()
```
